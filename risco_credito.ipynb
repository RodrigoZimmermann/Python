```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

data_risc_cred = open('dataset_risco_credito.csv')

data = data_risc_cred.readlines()

dataset_risco_credito = []
for i in range(1, len(data)):
  e = data[i].replace('\n', '').split(',')
  dataset_risco_credito.append(e)

print(np.matrix(dataset_risco_credito))
```

```python
from sklearn.preprocessing import LabelEncoder
import pickle

X_risco_credito = []
y_risco_credito = []

historia = []
divida = []
garantias = []
renda = []

temp = []
for i in dataset_risco_credito:
    historia.append(i[0])
    divida.append(i[1])
    garantias.append(i[2])
    renda.append(i[3])
    y_risco_credito.append(i[4])

label_encoder_historia = LabelEncoder()
label_encoder_divida = LabelEncoder()
label_encoder_garantias = LabelEncoder()
label_encoder_renda = LabelEncoder()

historia = label_encoder_historia.fit_transform(historia)
divida = label_encoder_divida.fit_transform(divida)
garantias = label_encoder_garantias.fit_transform(garantias)
renda = label_encoder_renda.fit_transform(renda)

for i in range(len(dataset_risco_credito)):
  X_risco_credito.append([historia[i], divida[i], garantias[i], renda[i]])

with open('risco_credito.pkl', 'wb') as f:
  pickle.dump([X_risco_credito, y_risco_credito], f)
```

```python
from sklearn.naive_bayes import GaussianNB

naiveb_risco_credito = GaussianNB()
naiveb_risco_credito.fit(X_risco_credito, y_risco_credito)
print(naiveb_risco_credito.predict([[0, 0, 1, 2]]))
print(naiveb_risco_credito.predict([[2, 0, 0, 0]]))
print(naiveb_risco_credito.classes_)
print(naiveb_risco_credito.class_count_)
```
